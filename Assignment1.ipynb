```python
#part a
import numpy as np

a = np.array([5, 2])
b = np.array([-3, 4])
```

```python
#part b

def dot(a, b):
    dot = 0
    for i in range(np.size(a)):
        dot += a[i]*b[i]
    return dot

print("Dot product: ", dot(a, b))
```

```python
#part c
import math
distance_sqr = 0

for i in range(np.size(a)):
    distance_sqr += (a[i] - b[i]) ** 2

e_dist = math.sqrt(distance_sqr)

print("Euclidean distance: ", e_dist)
```

```python
#part d
#beta = y dot xhat, projection = beta(xhat)

#find unit vector of b (b_hat)
b_norm_sqr = 0
for i in range(np.size(b)):
    b_norm_sqr += b[i] ** 2

b_norm = math.sqrt(b_norm_sqr)
b_hat = b / b_norm

beta = dot(a, b_hat)
#compute the projection of a (beta) onto b
projection  = beta * b_hat
print("Projection: ", projection)
```
