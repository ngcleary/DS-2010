```python
# Matrix A
import numpy as np
A = np.array([[2, 0], [0, 5]])

print(A)
```

```python
# Part a
# Eigenvalues of A are the diagonal since A is already diaganol matrix

eigen_values, eigen_vectors = np.linalg.eig(A)
print("Eigen Values: ", eigen_values)
print("Eigen Vectors: \n", eigen_vectors)

```

```python
# Part b - Diagonalize ğ´ using ğ´ = ğ‘„ğ›¬ğ‘„âˆ’1.

Q = eigen_vectors
Lambda = np.array([[eigen_values[0], 0], [0, eigen_values[1]]])
Q_inverse =  np.linalg.inv(Q)

print("Q: \n", Q)
print("Lambda: \n", Lambda)
print("Q_inverse: \n", Q_inverse)
```

```python
# Part c - Verify

A = Q @ Lambda @ Q_inverse
print("A = ğ‘„ğ›¬3ğ‘„âˆ’1: \n", A)
```

```python
# Part d - Compute A3 using ğ‘„ğ›¬3ğ‘„âˆ’1
# QQ-1 = I, so A3 = ğ›¬3

A3 = Lambda @ Lambda @ Lambda
print("A3 using ğ‘„ğ›¬3ğ‘„âˆ’1: \n", A3)
print("A3 (A * A * A): \n", A @ A @ A)
```
